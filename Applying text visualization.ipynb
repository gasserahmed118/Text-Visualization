# Imports and Model Loading

```python
import spacy
from spacy import displacy
from IPython.core.display import display, HTML

# Load English model
nlp_en = spacy.load('en_core_web_sm')

# Load Arabic / multilingual model (optional)
# nlp_ar = spacy.load('xx_ent_wiki_sm')
```

# Dependency Parsing Visualization (English)

```python
text = u'Apple is going to build a U.K. factory for $6 million.'
doc = nlp_en(text)

# Render dependency parse
displacy.render(doc, style='dep', jupyter=True, options={'distance': 90})
print ("------------------------------------------------------------------------------------")

# Print token details
for token in doc:
    print(f'{token.text:{12}} {token.pos_:{8}} {token.dep_:{10}} {spacy.explain(token.dep_)}')
```

# Named Entity Visualization (English)

```python
text = u'Over the last quarter Apple sold nearly 20 thousand iPods for a profit of $6 million. By contrast, Sony sold only 7 thousand Walkman music players.'
doc = nlp_en(text)

# Standard entity visualization
displacy.render(doc, style='ent', jupyter=True)

# Customized colors for specific entity types
colors = {'ORG': 'linear-gradient(90deg, #aa9cfc, #fc9ce7)',
          'PRODUCT': 'radial-gradient(yellow, green)',
          'MONEY': 'linear-gradient(orange, red)'}
options = {'ents': ['ORG', 'PRODUCT', 'MONEY'], 'colors': colors}

displacy.render(doc, style='ent', jupyter=True, options=options)
```

# Sentence-Level Rendering for Long Texts


```python
long_text = """
In ancient Rome, some neighbors live in three adjacent houses. 
In the center is the house of Senex, who lives there with wife Domina, son Hero, and several slaves. 
One day, Senex and Domina go on a trip and leave Pseudolus in charge of Hero.
"""

doc = nlp_en(long_text)

# Render each sentence separately for dependency and entity visualization
for sent in doc.sents:
    sent_doc = nlp_en(sent.text)
    print("Dependency parse:")
    displacy.render(sent_doc, style='dep', jupyter=True, options={'distance': 90})
    print("Named entities:")
    displacy.render(sent_doc, style='ent', jupyter=True)
```

# Dependency and Entity Visualization for Financial Text


```python
finance_text = """
Over the last quarter, Microsoft sold 10 thousand Surface laptops for $8 million. 
Apple sold 15 thousand iPhones for $12 million. 
Tesla earned $20 million in revenue from car sales.
"""

doc = nlp_en(finance_text)

# Render each sentence separately for clarity
for sent in doc.sents:
    sent_doc = nlp_en(sent.text)
    displacy.render(sent_doc, style='dep', jupyter=True, options={'distance': 80})
    displacy.render(sent_doc, style='ent', jupyter=True)
```

# Dependency and Entity Visualization for Historical Text


```python
history_text = """
Silver coins in the Roman Empire were debased over time. 
The government mixed copper and lead with silver to create more coins, lowering their value. 
During the Malian king Mansa Musa's hajj to Mecca, he spent so much gold that its price fell for over a decade.
"""

doc = nlp_en(history_text)

# Sentence-level rendering
for sent in doc.sents:
    sent_doc = nlp_en(sent.text)
    displacy.render(sent_doc, style='dep', jupyter=True, options={'distance': 80})
    displacy.render(sent_doc, style='ent', jupyter=True)
```

# Manual Entity Annotation and Custom Colors


```python
text = "But Google is starting from behind."
doc = nlp_en(text)

# Manual entity input
manual_example = [{"text": text,
                   "ents": [{"start": 4, "end": 10, "label": "ORG"}],
                   "title": None}]

displacy.render(manual_example, style="ent", manual=True)

# Custom colors example
colors = {"ORG": "linear-gradient(90deg, #aa9cfc, #fc9ce7)"}
options = {"ents": ["ORG"], "colors": colors}
displacy.render(doc, style="ent", options=options)
```

# HTML visualization With English example

```python
import spacy
from spacy import displacy
from IPython.core.display import display, HTML

nlp = spacy.load("en_core_web_sm")

text_en = "Apple is going to build a U.K. factory for $6 million."
doc_en = nlp(text_en)

html_en = displacy.render(doc_en, style="dep", options={"distance": 90})
display(HTML(html_en))
```

# HTML visualization With Arabic example

```python
text_ar = "       5  "
doc_ar = nlp(text_ar)

html_ar = displacy.render(doc_ar, style="ent")
display(HTML(html_ar))
```

# HTML visualization With Arabic example

```python
text = """
           .     .     164 781 (  )     232   ( 847).                .                              .    850 /232                         .

       1145     .        Algebra  Zero                       .                 .

"""
doc = nlp_en(text)
sentence_spans = list(doc.sents)
html = displacy.render(sentence_spans, style="dep",options={'distance': 50})

display(HTML(html))
```

# Sentence-Level Arabic Rendering

```python
text_ar = """
     .
     .
    .
"""

doc = nlp_en(text_ar)

sentences = list(doc.sents)

html = displacy.render(
    sentences,
    style="dep",
    page=True,
    options={
        "distance": 70,
        "bg": "#00A3CF",
        "color": "white",
        "font": "Arial"
    }
)

display(HTML(html))
```

# Arabic Named Entity Recognition (NER)

```python
text_ar = "          6  ."
doc = nlp_en(text_ar)

html = displacy.render(
    doc,
    style="ent",
    options={
        "ents": ["ORG", "GPE", "MONEY"],
        "colors": {
            "ORG": "linear-gradient(90deg, #aa9cfc, #fc9ce7)",
            "GPE": "#7aecec",
            "MONEY": "#feca74"
        }
    }
)

display(HTML(html))
```
